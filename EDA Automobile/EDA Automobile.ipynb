# Most expensive car according to price

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')     #replace any non-numeric values (e.g., strings like '?', 'unknown', etc.) with NaN
```

```python
df = df.dropna(subset=['price'])               # dropna(): Removes rows with missing (NaN) values.
                                               #subset=['price']: Only considers missing values in the 'price' column.
```

```python
most_expensive = df.loc[df['price'].idxmax()]    # .idxmax() â†’ returns the index (row number) of the maximum value in the 'price' column.
                                                 # df.loc[...] â†’ accesses the full row at that index using .loc.
```

```python
print("Company:", most_expensive['make'])
print("Price: $", most_expensive['price'])
```

```python
# Create a new column which stores the number of doors in a car as integers.
df['num-of-doors'] = df['num-of-doors'].replace({'four': 4, 'two': 2, '?': None})
```

# Qh) Calculate which variable/feature/attribute is impacting the price of the car the most.

```python
df.columns
```

```python
# Convert price and other numeric columns to numbers
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['engine-size'] = pd.to_numeric(df['engine-size'], errors='coerce')
df['curb-weight'] = pd.to_numeric(df['curb-weight'], errors='coerce')
```

```python
# Drop rows with missing values in these columns
df = df.dropna(subset=['price', 'horsepower', 'engine-size', 'curb-weight'])
```

```python
# Check correlation between price and other columns
correlations = df[['price', 'horsepower', 'engine-size', 'curb-weight']].corr()
```

```python
print("Correlation with price:")
print(correlations['price'].sort_values(ascending=False))
```

# Qi) Concatenate the two data frames given below firstly row wise and secondly column wise

```python
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845,
171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995,
23600, 61500 , 58900]}

german = pd.DataFrame(GermanCars)
japanese = pd.DataFrame(japaneseCars)

row_concat = pd.concat([german, japanese], ignore_index=True)
print(row_concat)
```

```python
col_wise = pd.concat([german, japanese], axis=1)
print("\nColumn-wise Concatenation:")
print(col_wise)
```
